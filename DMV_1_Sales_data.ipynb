```python
import pandas as pd

# Load CSV
csv_data = pd.read_csv('sales_data_sample.csv', encoding='latin-1')

# Load Excel
excel_data = pd.read_excel('sales_data_sample.xlsx')

# Load JSON
json_data = pd.read_json('csvjson.json')


```

```python
csv_info = csv_data.info()
excel_info = excel_data.info()
json_info = json_data.info()

csv_head = csv_data.head()
excel_head = excel_data.head()
json_head = json_data.head()

csv_info, excel_info, json_info, csv_head, excel_head, json_head
```

```python
# Handle missing values by filling or dropping them
# For simplicity, let's fill missing 'STATE' and 'TERRITORY' with "Unknown" and 'ADDRESSLINE2' with an empty string
csv_data['STATE'].fillna('Unknown', inplace=True)
csv_data['TERRITORY'].fillna('Unknown', inplace=True)
csv_data['ADDRESSLINE2'].fillna('', inplace=True)

excel_data['STATE'].fillna('Unknown', inplace=True)
excel_data['TERRITORY'].fillna('Unknown', inplace=True)
excel_data['ADDRESSLINE2'].fillna('', inplace=True)

json_data['STATE'].fillna('Unknown', inplace=True)
json_data['TERRITORY'].fillna('Unknown', inplace=True)
json_data['ADDRESSLINE2'].fillna('', inplace=True)
```

```python
# Drop duplicate records if any
csv_data.drop_duplicates(inplace=True)
excel_data.drop_duplicates(inplace=True)
json_data.drop_duplicates(inplace=True)

```

```python
# Concatenate all three DataFrames into one unified DataFrame
combined_data = pd.concat([csv_data, excel_data, json_data], ignore_index=True)

# Display info and head of the combined data to confirm the integration and cleaning
combined_data_info = combined_data.info()
combined_data_head = combined_data.head()

combined_data_info, combined_data_head
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming `combined_data` is the merged DataFrame

# Step 5: Data Transformation
# Split 'ORDERDATE' into 'ORDER_YEAR' and 'ORDER_MONTH' for time-based analysis
combined_data['ORDERDATE'] = pd.to_datetime(combined_data['ORDERDATE'], errors='coerce')
combined_data['ORDER_YEAR'] = combined_data['ORDERDATE'].dt.year
combined_data['ORDER_MONTH'] = combined_data['ORDERDATE'].dt.month

# Derive a new variable 'TOTAL_ORDER_VALUE'
combined_data['TOTAL_ORDER_VALUE'] = combined_data['QUANTITYORDERED'] * combined_data['PRICEEACH']

# Step 6: Descriptive Statistics and Aggregation
# Total Sales
total_sales = combined_data['SALES'].sum()

# Average Order Value
average_order_value = combined_data['TOTAL_ORDER_VALUE'].mean()

# Product Category Distribution
product_distribution = combined_data['PRODUCTLINE'].value_counts()
```

```python
# Step 7: Visualizations
# Bar Plot for Total Sales by Product Line
plt.figure(figsize=(10, 6))
sns.barplot(data=combined_data, x='PRODUCTLINE', y='SALES', estimator=sum, ci=None)
plt.title('Total Sales by Product Line')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie Chart for Sales Distribution by Country
plt.figure(figsize=(8, 8))
combined_data.groupby('COUNTRY')['SALES'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Distribution by Country')
plt.ylabel('')
plt.show()
```

```python
# Box Plot for Sales Distribution
plt.figure(figsize=(10, 6))
sns.boxplot(data=combined_data, y='SALES')
plt.title('Distribution of Sales')
plt.ylabel('Sales')
plt.show()
```
